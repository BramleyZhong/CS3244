```python
match_data['Date'] = pd.to_datetime(match_data['Date'])
match_data['Date'] = match_data['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
match_data['Date'] = match_data['Date'].apply(lambda x: x[0:4] + x[5:7] + x[8:10])
```

```python
def get_match_id(row):
    id = row['Date']+ '0'
    id += row['Tm'] 
    url = 'https://www.basketball-reference.com/boxscores/' + id + '.html'
    return url
```

```python
def get_player_data(url, match_id):
    try:
        # get first team team player
        tables = pd.read_html(url)
        df_guest = tables[0]
        df_guest.columns = ['Starters', 'MP', 'FG', 'FGA', 'FG%', '3P','3PA', '3P%', 'FT', 'FTA', 
                        'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', '+/-']
        df_guest = df_guest[df_guest['Starters'] != 'Reserves']
        
        df_guest_add = tables[7]
        df_guest_add.columns = ['Starters', 'MP', 'TS%', 'eFG%', '3PAr', 'FTr', 'ORB%','DRB%', 'TRB%', 
                            'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'ORtg', 'DRtg', 'BPM']
        df_guest_add = df_guest_add[df_guest_add['Starters'] != 'Reserves']

        df_guest_all = df_guest.merge(df_guest_add, on = ['Starters', 'MP'])
        

        # get second team player
        df_host = tables[8]
        df_host.columns = df_guest.columns
        df_host = df_host[df_host['Starters'] != 'Reserves']
        
        df_host_add = tables[15]
        df_host_add.columns = df_guest_add.columns
        df_host_add = df_host_add[df_host_add['Starters'] != 'Reserves']

        # combine two team
        df_host_all = df_host.merge(df_host_add, on = ['Starters', 'MP'])
        df_host_all.loc['end_of_host'] = np.nan
        result = pd.concat([df_host_all, df_guest_all], axis = 0)
        # result.to_csv('./players/' + match_id + '.csv', index = 0)
    
    except:
        pass
```

```python
for index, row in match_data.iterrows():
    match_id = row['Date'] + '_' + row['Tm'] + '_' + row['Opp']
    url = get_match_id(row)
    get_player_data(url, match_id)
```
