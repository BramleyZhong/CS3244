```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./match_player_data.csv")
df.head()
```

```python
# calculate A/T, efficiency(EFF) and EFFECTIVE FIELD GOAL PERCENTAGE 
df["A/T"] = df["AST_players"]/df["TOV_players"]
df["EFF"] = df["PTS_players"]+df["TRB_players"]+df["AST_players"]+df["STL_players"]+df["BLK_players"]
-(df["FGA_players"]-df["FG_players"])-(df["FTA_players"]-df["FT_players"])-df["TOV_players"]
df.head()
```
